1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f = 10 * x - 5

derivative_f = sp.diff(f, x)

print(f"The derivative of the function f(x) = 10x - 5 is: {derivative_f}")
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x_sym = sp.symbols('x')

f_sym = sp.sin(x_sym)

df_sym = sp.diff(f_sym, x_sym)

f_num = sp.lambdify(x_sym, f_sym, 'numpy')
df_num = sp.lambdify(x_sym, df_sym, 'numpy')

x = np.linspace(0, 2 * np.pi, 400)

f = f_num(x)
df = df_num(x)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = sin(x)', color='blue')
plt.plot(x, df, label="f'(x) = cos(x)", color='green')

x_point = np.pi / 4
f_point = f_num(x_point)
df_point = df_num(x_point)
tangent_line = df_point * (x - x_point) + f_point
plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label=f'Tangent at x = {x_point:.2f}')


plt.title("Graph of the Function and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x_sym = sp.symbols('x')
f_sym = sp.ln(x_sym)

df_sym = sp.diff(f_sym, x_sym)

f_num = sp.lambdify(x_sym, f_sym, 'numpy')
df_num = sp.lambdify(x_sym, df_sym, 'numpy')

# Є проблеми як раз з візуалізацією, важко дається розуміння саме на яких відрізках підображати функції
x = np.linspace(0.1, 5, 400)

f = f_num(x)
df = df_num(x)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = ln(x)', color='blue')
plt.plot(x, df, label="f'(x) = 1/x", color='green')

x_point = 2
f_point = f_num(x_point)
df_point = df_num(x_point)
tangent_line = df_point * (x - x_point) + f_point

plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label=f'Tangent at x = {x_point:.2f}')

plt.title("Graph of the Function and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x_sym = sp.symbols('x')

# Прописуємо функцію
f_sym = x_sym**2 - 5*x_sym + 1

# Розрахуємо похідну
df_sym = sp.diff(f_sym, x_sym)

# Конвертуємо символічні функції в числові
f_num = sp.lambdify(x_sym, f_sym, 'numpy')
df_num = sp.lambdify(x_sym, df_sym, 'numpy')

# Вказуємо відрізок
x = np.linspace(0, 100, 400)

# Розраховуємо функцію і її похідну
f = f_num(x)
df = df_num(x)

# Створюємо плот
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df, label="f'(x)", color='green')

# Додаємо позначки
plt.title("Graph of the Function and its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Відображаємо
plt.show()
```
